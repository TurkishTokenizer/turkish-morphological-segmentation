```python
# Heart of the tokenizer
root = pynini.union("kiler", "kitap", "ev", "okul")
plural = pynini.union("lar", "ler")
case = pynini.union("a", "e", "da", "de", "dan", "den")

turkish_word = root + plural.ques + case.ques
turkish_word.optimize()
```

```python
# Valid Tests
for s in ["kiler", "kilerler", "kitaplara", "evindeydiler", "evde", "okullardan", "okul"]:
    result = pynini.accep(s) @ turkish_word
    if result.num_states() > 0:
        print(f"{s}: ✅ valid")
    else:
        print(f"{s}: ❌ invalid")
```
